```python
# nn_qspr_nf_with_plots.py
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import LeaveOneOut
import tensorflow as tf
from tensorflow.keras import layers, regularizers, models, callbacks, optimizers
import matplotlib.pyplot as plt
import seaborn as sns
import os, random

# ============================================================================================================
# 1. Reproducibility setup

seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)
os.environ['PYTHONHASHSEED'] = str(seed)

# ============================================================================================================
# 2. Load dataset

df = pd.read_csv('nf_drugs_topo_qspr.csv')

FEATURES = ['M2', 'Randic', 'SD2', 'MIR', 'SZ', 'MH']
TARGET = 'MW'

mask = df[FEATURES + [TARGET]].notnull().all(axis=1)
df_sub = df.loc[mask].reset_index(drop=True)
print(f"Using {len(df_sub)} samples for target '{TARGET}'")

X = df_sub[FEATURES].astype(float).values
y = df_sub[TARGET].astype(float).values

# ============================================================================================================
# 3. Standardize features

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================================================================
# 4. Model builder

def build_model(input_dim, lr=1e-3, l2=1e-4, dropout_rate=0.2):
    inp = layers.Input(shape=(input_dim,))
    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(l2))(inp)
    x = layers.Dropout(dropout_rate)(x)
    x = layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(l2))(x)
    x = layers.Dropout(dropout_rate)(x)
    out = layers.Dense(1, activation='linear')(x)
    model = models.Model(inputs=inp, outputs=out)
    model.compile(optimizer=optimizers.Adam(learning_rate=lr), loss='mse')
    return model

# ============================================================================================================
# 5. Leave-One-Out CV

loo = LeaveOneOut()
y_true, y_pred = [], []
results = []
fold = 0

for train_index, test_index in loo.split(X_scaled):
    fold += 1
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = build_model(input_dim=X_scaled.shape[1])
    es = callbacks.EarlyStopping(monitor='val_loss', patience=80, restore_best_weights=True, verbose=0)
    history = model.fit(
        X_train, y_train,
        validation_split=0.2 if len(X_train) > 4 else 0.25,
        epochs=1000,
        batch_size=4,
        callbacks=[es],
        verbose=0
    )

    yhat = model.predict(X_test).ravel()
    y_true.extend(y_test.ravel().tolist())
    y_pred.extend(yhat.tolist())

    rmse = np.sqrt(mean_squared_error(y_test, yhat))
    mae = mean_absolute_error(y_test, yhat)
    results.append({'fold': fold, 'y_true': float(y_test[0]), 'y_pred': float(yhat[0]), 'rmse': rmse, 'mae': mae})

    print(f"Fold {fold:02d}: true={float(y_test):.4f}, pred={float(yhat):.4f}, rmse={rmse:.4f}, mae={mae:.4f}")

# ============================================================================================================
# 6. Global Metrics

global_rmse = np.sqrt(mean_squared_error(y_true, y_pred))
global_mae = mean_absolute_error(y_true, y_pred)
global_r2 = r2_score(y_true, y_pred) if len(y_true) > 1 else float('nan')

print("\nLOO CV summary:")
print(f"Samples: {len(y_true)}")
print(f"Mean RMSE across folds: {np.mean([r['rmse'] for r in results]):.4f} ± {np.std([r['rmse'] for r in results]):.4f}")
print(f"Mean MAE  across folds: {np.mean([r['mae'] for r in results]):.4f} ± {np.std([r['mae'] for r in results]):.4f}")
print(f"Global RMSE (all preds): {global_rmse:.4f}")
print(f"Global MAE  (all preds): {global_mae:.4f}")
print(f"Global R²   (all preds): {global_r2:.4f}")

# ============================================================================================================
# 7. Save predictions

out_df = pd.DataFrame(results)
out_df.to_csv(f'predictions_{TARGET}.csv', index=False)
print(f"\nPredictions saved to predictions_{TARGET}.csv")

# ============================================================================================================
# 8. Visualization & Performance Tables

df_perf = pd.DataFrame({
    "Metric": ["Mean RMSE (±SD)", "Mean MAE (±SD)", "Global RMSE", "Global MAE", "Global R²"],
    "Value": [
        f"{np.mean([r['rmse'] for r in results]):.3f} ± {np.std([r['rmse'] for r in results]):.3f}",
        f"{np.mean([r['mae'] for r in results]):.3f} ± {np.std([r['mae'] for r in results]):.3f}",
        f"{global_rmse:.3f}",
        f"{global_mae:.3f}",
        f"{global_r2:.3f}"
    ]
})
print("\n=== Model Performance Summary ===")
print(df_perf.to_string(index=False))

# Actual vs Predicted Table

df_ap = pd.DataFrame({"Actual": y_true, "Predicted": y_pred, "Residual": np.array(y_pred) - np.array(y_true)})
print("\n=== Actual vs Predicted Values ===")
print(df_ap.head(10))

# ===============================================================================================================
# 9. Plots

plt.figure(figsize=(6,6))
sns.regplot(x=y_true, y=y_pred, ci=None, line_kws={"color":"red"})
plt.xlabel("Actual MW")
plt.ylabel("Predicted MW")
plt.title("Predicted vs Actual")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
residuals = np.array(y_pred) - np.array(y_true)
plt.scatter(y_true, residuals, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Actual MW")
plt.ylabel("Residual (Pred - Actual)")
plt.title("Residual Plot")
plt.grid(True)
plt.show()

print("\n All plots and performance summaries generated successfully.")
```
